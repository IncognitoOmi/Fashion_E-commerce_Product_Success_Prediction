```python
import pandas as pd
from pycaret.classification import *
```

```python
df = pd.read_csv('/content/prediction_input.csv')
df.columns
```

```python
class ProductSuccessPrediction:
    def __init__(self, prediction_data_path):
        self.prediction_data_path = prediction_data_path
        self.model = None

    def load(self):
        # Load prediction data
        self.prediction_data = pd.read_csv(self.prediction_data_path)
        return self.prediction_data

    def preprocess(self):
        return self.prediction_data

    def load_model(self, model_name='best_model'):
        # Load the saved model
        self.model = load_model(model_name)

    def predict(self):
        # Preprocess and make predictions
        preprocessed_data = self.preprocess()
        predictions = predict_model(self.model, data=preprocessed_data)

        return predictions

#actual usage
prediction_data_path = '/content/prediction_input.csv'
model_selector = ProductSuccessPrediction(prediction_data_path)
model_selector.load()
model_selector.load_model(model_name='best_model')
predictions = model_selector.predict()

# Display the final output with 'success_indicator'

predictions.head()
```

```python
predictions.to_csv('prediction_output.csv', index=False)
```

```python
import pandas as pd
from pycaret.classification import *

class ProductSuccessTraining:
    def __init__(self, historic_data_path):
        self.historic_data_path = historic_data_path
        self.model = None

    def load(self):
        # Load historic data
        self.historic_data = pd.read_csv(self.historic_data_path)
        return self.historic_data

    def preprocess(self):
        # Remove unnecessary columns
        return self.historic_data.drop(columns=['item_no'])

    def train(self, target_column):
        # Setup PyCaret environment
        preprocessed_data = self.preprocess()
        clf = setup(data=preprocessed_data, target=target_column, session_id=123)
        # Train and select the best model
        self.model = compare_models()
        return self.model

    def save_model(self, model_name='best_model'):
        # Save the trained model
        save_model(self.model, model_name)

# Example usage:
historic_data_path = '/content/historic.csv'
model_trainer = ProductSuccessTraining(historic_data_path)
model_trainer.load()
best_model = model_trainer.train(target_column='success_indicator')
print(best_model)
```
